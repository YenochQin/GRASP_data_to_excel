x，y：长度相同的数组，也就是我们即将绘制散点图的数据点，输入数据。

s：点的大小，默认 20，也可以是个数组，数组每个参数为对应点的大小。

c：点的颜色，默认蓝色 'b'，也可以是个 RGB 或 RGBA 二维行数组。

marker：点的样式，默认小圆圈 'o'。

cmap：Colormap，默认 None，标量或者是一个 colormap 的名字，只有 c 是一个浮点数数组的时才使用。如果没有申明就是 image.cmap。

norm：Normalize，默认 None，数据亮度在 0-1 之间，只有 c 是一个浮点数的数组的时才使用。

vmin，vmax：：亮度设置，在 norm 参数存在时会忽略。

alpha：：透明度设置，0-1 之间，默认 None，即不透明。

linewidths：：标记点的长度。

edgecolors：：颜色或颜色序列，默认为 'face'，可选值有 'face', 'none', None。

plotnonfinite：：布尔值，设置是否使用非限定的 c ( inf, -inf 或 nan) 绘制点。

**kwargs：：其他参数。

````python
from fig_size import set_size

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import rcParams
config = {
    "font.family":'Times New Roman',
    "font.size": 14
}
rcParams.update(config)
````

````python
fig, ax = plt.subplots(1, 1, figsize=set_size(426.79135))
ax = plt.gca()
ax.tick_params(which='both', direction='in')

props = {'xlabel': 'Level Number','ylabel':r"$(\tau_l - \tau_v)/\tau_l$"} # 坐标系-坐标轴-刻度-标签
ax.set(**props)

ax.set_xlim(left=0., right=len(lifetime)+1)
ax.set_xticks(np.arange(0,len(lifetime)+1,10))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(5))

# ax.set_yticks(np.arange(-0.2,0.8,0.2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))

com_color = {'red' : '#ff0000', 'yellow' : '#e99e11', 'green' : '#2dc431', 'dcyan' : '#2daea8', 'blue' : '#3468ca', 'Purple' : '#7727b5', 'pink' : '#d22b82', 'grey' : '#6d6c6d'}

ax.scatter(lifetime['No'], lifetime['Cu'], marker="x", s=12, linewidths=1)
ax.scatter(lifetime['No'], lifetime['Zn'], marker="x", s=12, linewidths=1)
ax.scatter(lifetime['No'], lifetime['Ga'], marker="x", s=12, linewidths=1)
ax.scatter(lifetime['No'], lifetime['Ge'], marker="x", s=12, linewidths=1)
ax.scatter(lifetime['No'], lifetime['As'], marker="x", s=12, linewidths=1)
ax.scatter(lifetime['No'], lifetime['Se'], marker="x", s=12, linewidths=1)
ax.scatter(lifetime['No'], lifetime['Br'], marker="x", s=12, linewidths=1)

ax.legend(('Cu', 'Zn', 'Ga', 'Ge', 'As', 'Se', 'Br'),frameon=False,ncol=2,loc='best', bbox_to_anchor=(0.7, 0.5, 0.3, 0.5),handlelength=0.08,  columnspacing=1.2) 
fig.savefig('D:\\LaTeX\\lanthanide_calculation\\data\\fig\\lifetime_differ.pdf', format='pdf', bbox_inches='tight')
````
